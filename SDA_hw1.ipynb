```python
#60212407 이진서 hw1
import numpy as np
import matplotlib.pyplot as plt
import time
import seaborn as sns


row= [10, 100, 1000]
column = [10, 100, 1000,10000]


for a in row:
    for b in column:
        time_loop = []
        time_vect = []

        #for loop 방법
        for _ in range(5):
            start_time=time.time()
            matrix_loop = np.zeros((a,b))
            for i in range(a):
                for j in range(b):
                    matrix_loop[i,j] = np.sqrt((i-j)**2) #유클리드
            end_time=time.time()
            time_loop.append(end_time-start_time)

            #vectorize 방법
            start_time=time.time()
            matrix_vect = np.sqrt((np.arange(a)[:,np.newaxis] - np.arange(b))**2)
            end_time=time.time()
            time_vect.append(end_time-start_time)

        print(f"for loop: <row ={a} and column= {b}>, 실행 시간 : {time_loop}")
        print(f"vectorize: <row={a} and column={b}>, 실행 시간: {time_vect}\n")

print("For loop distance matrix:")
print(matrix_loop)

print("Vectorize distance matrix:")
print(matrix_vect)

#plot
fig, axs=plt.subplots(nrows=1, ncols=2, figsize=(14,6))

sns.heatmap(matrix_loop,cmap='Blues',ax=axs[0])
axs[0].set_title('Distance Matrix (For Loop)')

sns.heatmap(matrix_vect,cmap='Blues',ax=axs[1])
axs[1].set_title('Distance Matrix (Vectorization)')

plt.show()
```
